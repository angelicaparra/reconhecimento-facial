```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")  # modelo base

results = model.predict(
    source="angelinajolie.jpg",
    conf=0.5,
    classes=[0]  # classe 0 = pessoa, detecta rosto como região superior
)
```
